```python
%%javascript


const numbers = [1, 2, 3, 4, 5];


// Print the original array
console.log("Original array:", numbers);


// Access elements by index
const firstElement = numbers[0];
const thirdElement = numbers[2];
console.log("First element:", firstElement);
console.log("Third element:", thirdElement);


// Modify elements in the array
numbers[1] = 10;
console.log("Modified array:", numbers);


// Add elements to the end of the array
numbers.push(6);
console.log("Array with added element:", numbers);


// Remove elements from the end of the array
const removedElement = numbers.pop();
console.log("Removed element:", removedElement);
console.log("Array after removing element:", numbers);


// Find the length of the array
const arrayLength = numbers.length;
console.log("Array length:", arrayLength);


// Iterate through the array
console.log("Iterating through the array:");
for (let i = 0; i < numbers.length; i++) {
   console.log(numbers[i]);
}


// Use forEach to iterate through the array
console.log("Using forEach:");
numbers.forEach((number) => {
   console.log(number);
});


// Filter the array to select even numbers
const evenNumbers = numbers.filter((number) => number % 2 === 0);
console.log("Even numbers:", evenNumbers);
```
